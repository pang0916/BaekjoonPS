```python
import sys
from heapq import heappop, heappush
input = sys.stdin.readline
INF = 500001

def makeEdge(nodeSize, edgeSize):
    Edges = [[] for _ in range(nodeSize)]
    isEdges = [[False]*nodeSize for _ in range(nodeSize)]
    for i in range(edgeSize):
        curr, next, cost = map(int, input().split())
        Edges[curr].append([cost, next])
        isEdges[curr][next] = True
    return Edges, isEdges

def delEdges():
    q = [endNode]
    while q:
        curr = q.pop()
        for next in Prevs[curr]:
            if isEdges[next][curr]:
                isEdges[next][curr] = False
                q.append(next)

while True:
    nodeSize, edgeSize = map(int, input().split())
    if nodeSize == 0 and edgeSize == 0:
        break
    
    startNode, endNode = map(int, input().split())
    Edges, isEdges = makeEdge(nodeSize, edgeSize)
    
    #dijkstra first
    Costs = [INF for _ in range(nodeSize)]
    Prevs = [[] for _ in range(nodeSize)]
    que = [[0, startNode]]
    while que:
        cost, curr = heappop(que)
        if Costs[curr] < cost:
            continue
        if Costs[endNode] < cost:
            break
        
        Costs[curr] = cost
        for cost, next in Edges[curr]:
            if Costs[next] == Costs[curr]+cost :
                Prevs[next].append(curr)
            elif Costs[next] > Costs[curr]+cost:
                Costs[next] = Costs[curr]+cost
                Prevs[next] = [curr]
                heappush(que, [Costs[next], next])
    
    delEdges()

    #dijkstra second
    Costs = [INF for _ in range(nodeSize)]
    que = [[0, startNode]]
    while que:
        cost, curr = heappop(que)
        
        if Costs[curr] < cost:
            continue
        
        Costs[curr] = cost
        if curr == endNode:
            break
        
        for cost, next in Edges[curr]:
            if isEdges[curr][next] and Costs[next] > Costs[curr]+cost:
                Costs[next] = Costs[curr]+cost
                heappush(que, [Costs[next], next])
                
    print(Costs[endNode]) if Costs[endNode] != INF else print(-1)
```
